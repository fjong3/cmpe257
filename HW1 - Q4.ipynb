```python
from sklearn import tree
```

```python
#4a Process the data and make it available for tree classfier
"""
Data Processing documentation
X[0] = Marital:
Single = 0
Married = 1
Divorced = 2

X[1] = Estate:
No = 0
Yes = 1

X[2] Income:
(Income <= 75) = 0
(75 < Income < 100) = 1 
(Income <= 100) = 2

Evade:
No = 0
Yes = 1
"""
X = [
    [0, 1, 2],
    [1, 0, 2],
    [0, 0, 0],
    [1, 1, 2],
    [2, 0, 1],
    [1, 0, 0],
    [2, 1, 2],
    [0, 0, 1],
    [1, 0, 0],
    [0, 0, 1],
]
Y = [0, 0, 0, 0, 1, 0, 0, 1, 0, 1]
```

```python
# Generate the tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
# Feature names and class names
debt_features = ["Marital", "Estate", "Income"]
debt_classes = ["No Evade", "Evade"]
# Graph the tree
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None,
                               feature_names = debt_features,
                               class_names = debt_classes,
                               filled = True, rounded = True,
                               ) 
graph = graphviz.Source(dot_data) 
#graph.render("Debt Evasion") 
graph
```
